# Seaborn

Seaborn is a Python data visualization library based on Matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics. Here are some of the key features and functionalities of Seaborn:

1. **Built-in themes**: Seaborn comes with several built-in themes that make it easy to switch between different visual styles.

2. **Complex visualizations**: Seaborn simplifies the creation of complex visualizations like heatmaps, time series plots, and violin plots.

3. **Statistical aggregation**: Seaborn can handle statistical aggregation to summarize data, making it easier to display data distributions and trends.

4. **Integration with Pandas**: Seaborn works well with Pandas data structures, making it easy to visualize data stored in DataFrames.

5. **Color palettes**: Seaborn provides a variety of color palettes for styling plots, which can be customized to suit the specific requirements of your visualizations.

### Basic Example

Here’s a simple example of how to use Seaborn to create a basic plot:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load an example dataset
data = sns.load_dataset("tips")

# Create a simple scatter plot
sns.scatterplot(x="total_bill", y="tip", data=data)

# Show the plot
plt.show()
```

### Advanced Example

Creating a more complex plot with Seaborn, such as a heatmap:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load an example dataset
data = sns.load_dataset("flights")

# Pivot the data to create a matrix format
data_pivot = data.pivot("month", "year", "passengers")

# Create a heatmap
sns.heatmap(data_pivot, annot=True, fmt="d", cmap="YlGnBu")

# Show the plot
plt.show()
```

These examples demonstrate Seaborn’s ability to handle different types of data and create visually appealing plots with minimal code.

## Histogram (sns.histplot)
Description: A histogram is used to visualize the distribution of a single continuous variable by dividing the data into bins and 
counting the number of observations in each bin.

x: Variable to plot on the x-axis.

hue: Variable in data to map plot aspects to different colors.

binrange: Lower and upper range of bins.

stat: Aggregate statistic to compute in each bin (e.g., "count", "density").

common_bins: Use the same bins for all subsets of data.

common_norm: Normalize across all subsets.

multiple: Display multiple distributions ("layer", "dodge", "stack", "fill").

element: Plot element type ("bars", "step", "poly").

fill: Whether to fill the plot.

kde: Whether to plot a kernel density estimate.

````python
data=sns.load_dataset("tips")
data
````

````python
sns.histplot(data=data, x="day",hue="sex",kde=True)
plt.plot()
````

````python
data=sns.load_dataset("titanic")
sns.histplot(data=data, x="age",hue="who",kde=True,discrete=True) #discrete for closely packed data
plt.plot()
````

## Heatmap (sns.heatmap)
Description: Used to display data as a matrix of color-coded cells.
Parameters:
data: 2D dataset that can be coerced into an ndarray.
annot: If True, write the data value in each cell.
fmt: String formatting code to use when adding annotations.
cmap: Colormap to use for the plot.
linewidths: Width of the lines that will divide each cell.

data.pivot: A Pandas method to reshape the DataFrame.
index="month": The rows of the pivot table will be the months.
columns="year": The columns of the pivot table will be the years.
values="passengers": The cell values in the pivot table will be the number of passengers.
data_pivot: The resulting pivoted DataFrame where rows are months, columns are years, and the cells contain the passenger numbers.

sns.heatmap: A Seaborn function to create a heatmap, which is a graphical representation of data where values are represented with colors.
data_pivot: The pivoted DataFrame used to generate the heatmap.
annot=True: Annotates each cell with the numeric value.
fmt="d": Formats the annotation as an integer.
cmap="YlGnBu": Specifies the colormap to use (Yellow-Green-Blue in this case).

````python

# Load an example dataset
data = sns.load_dataset("flights")

# Pivot the data to create a matrix format
data_pivot = data.pivot(index="year", columns="month", values="passengers")

# Create a heatmap
sns.heatmap(data_pivot, annot=True, fmt="d", cmap="YlGnBu")

# Show the plot
plt.show()

````

### Strip Plot (`sns.stripplot`)

**Description**: A strip plot is used to display the distribution of a single categorical variable. It draws a scatterplot where one of the variables is categorical, which helps visualize the distribution of data points within each category.

**Parameters**:
- **`data`**: Dataset for the plot.
- **`x`**: Variable to plot on the x-axis (usually categorical).
- **`y`**: Variable to plot on the y-axis (usually continuous).
- **`hue`**: Variable in data to map plot aspects to different colors.
- **`jitter`**: Amount of jitter (only along the categorical axis) to apply.
- **`dodge`**: When using `hue`, whether to separate the strips for different levels of the hue variable.
- **`palette`**: Colors to use for the different levels of the hue variable.
- **`size`**: Size of the points.
- **`orient`**: Orientation of the plot (either 'v' for vertical or 'h' for horizontal).

### Example Usage

1. **Basic Strip Plot**:

   ```python
   import seaborn as sns
   import matplotlib.pyplot as plt

   # Load an example dataset
   data = sns.load_dataset("tips")

   # Create a strip plot
   sns.stripplot(x="day", y="total_bill", data=data)

   # Show the plot
   plt.show()
   ```

2. **Strip Plot with Jitter**:

   ```python
   sns.stripplot(x="day", y="total_bill", data=data, jitter=True)
   plt.show()
   ```

3. **Strip Plot with Hue**:

   ```python
   sns.stripplot(x="day", y="total_bill", hue="sex", data=data, palette="Set2")
   plt.show()
   ```

4. **Strip Plot with Dodge**:

   ```python
   sns.stripplot(x="day", y="total_bill", hue="sex", data=data, dodge=True)
   plt.show()
   ```

5. **Horizontal Strip Plot**:

   ```python
   sns.stripplot(y="day", x="total_bill", data=data, jitter=True)
   plt.show()
   ```

### Explanation of Parameters

- **`data`**: The dataset to plot.
- **`x`**: The column name of the data you want to plot on the x-axis (usually a categorical variable).
- **`y`**: The column name of the data you want to plot on the y-axis (usually a continuous variable).
- **`hue`**: Splits the data by the specified variable and color-codes them.
- **`jitter`**: Adds random noise to the data points to prevent overlapping. Useful for visualizing the distribution of data points within each category.
- **`dodge`**: When using the `hue` parameter, this separates the strips for different levels of the hue variable.
- **`palette`**: Specifies the colors to use for the different levels of the hue variable.
- **`size`**: Specifies the size of the points.
- **`orient`**: Specifies the orientation of the plot (vertical or horizontal).

### Use Cases
- **Visualizing Categorical Data**: Strip plots are great for showing the distribution of data points within different categories.
- **Comparing Distributions**: By using the `hue` parameter, you can compare the distributions of different subsets of the data within the same categories.
- **Adding Jitter**: Adding jitter helps in visualizing dense data points by spreading them out, making the distribution clearer.

### Example Scenarios

1. **Comparing Tips Given on Different Days**:

   ```python
   sns.stripplot(x="day", y="tip", data=data)
   plt.show()
   ```

2. **Comparing Total Bills by Day and Sex**:

   ```python
   sns.stripplot(x="day", y="total_bill", hue="sex", data=data, dodge=True)
   plt.show()
   ```

3. **Visualizing the Distribution of Total Bills**:

   ```python
   sns.stripplot(x="total_bill", data=data)
   plt.show()
   ```

By using the strip plot, you can effectively visualize the distribution and spread of data points within categorical variables, and gain insights into patterns and trends within your dataset.

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Load an example dataset
data = sns.load_dataset("tips")

# Create a strip plot
sns.stripplot(x="day", y="total_bill", data=data)

# Show the plot
plt.show()
````

### Categorical Plot (`sns.catplot`)

**Description**: `sns.catplot` is a versatile function in Seaborn for creating various types of categorical plots. It combines the functionality of several other categorical plots, such as strip plots, box plots, and bar plots, into one unified interface. This makes it easier to create complex visualizations with consistent styling.

**Parameters**:
- **`data`**: Dataset for the plot.
- **`x`**: Variable to plot on the x-axis (categorical).
- **`y`**: Variable to plot on the y-axis (usually continuous).
- **`hue`**: Variable in data to map plot aspects to different colors.
- **`kind`**: Type of plot to draw (e.g., "strip", "swarm", "box", "violin", "boxen", "point", "bar", "count").
- **`col`**: Variable to create subplot columns.
- **`row`**: Variable to create subplot rows.
- **`palette`**: Colors to use for the different levels of the hue variable.
- **`height`**: Height of each subplot.
- **`aspect`**: Aspect ratio of each subplot.
- **`order`**: Order to plot the categorical levels on the x-axis.
- **`col_order`**: Order to plot the categorical levels on the columns.
- **`row_order`**: Order to plot the categorical levels on the rows.

### Example Usage

1. **Basic Cat Plot (Strip Plot)**:

   ```python
   import seaborn as sns
   import matplotlib.pyplot as plt

   # Load an example dataset
   data = sns.load_dataset("tips")

   # Create a categorical plot with kind="strip"
   sns.catplot(x="day", y="total_bill", kind="strip", data=data)

   # Show the plot
   plt.show()
   ```

2. **Swarm Plot**:

3. **Box Plot**:

   ```python

   plt.show()
   ```

4. **Violin Plot**:

   ```python
   sns.catplot(x="day", y="total_bill", kind="violin", data=data)
   plt.show()
   ```

5. **Boxen Plot**:

   ```python
   sns.catplot(x="day", y="total_bill", kind="boxen", data=data)
   plt.show()
   ```

6. **Point Plot**:

   ```python
   sns.catplot(x="day", y="total_bill", kind="point", data=data)
   plt.show()
   ```

7. **Bar Plot**:

   ```python
   sns.catplot(x="day", y="total_bill", kind="bar", data=data)
   plt.show()
   ```

8. **Count Plot**:

   ```python

   plt.show()
   ```

9. **Facet Grid (Multiple Subplots)**:

   ```python
   sns.catplot(x="day", y="total_bill", hue="sex", col="time", kind="swarm", data=data)
   plt.show()
   ```

### Explanation of Parameters

- **`data`**: The dataset to plot.
- **`x`**: The column name of the data you want to plot on the x-axis (usually a categorical variable).
- **`y`**: The column name of the data you want to plot on the y-axis (usually a continuous variable).
- **`hue`**: Splits the data by the specified variable and color-codes them.
- **`kind`**: Specifies the type of plot to draw. Options include "strip", "swarm", "box", "violin", "boxen", "point", "bar", and "count".
- **`col`**: Creates subplot columns based on the specified variable.
- **`row`**: Creates subplot rows based on the specified variable.
- **`palette`**: Specifies the colors to use for the different levels of the hue variable.
- **`height`**: Specifies the height of each subplot.
- **`aspect`**: Specifies the aspect ratio of each subplot.
- **`order`**: Specifies the order to plot the categorical levels on the x-axis.
- **`col_order`**: Specifies the order to plot the categorical levels on the columns.
- **`row_order`**: Specifies the order to plot the categorical levels on the rows.

### Use Cases

1. **Visualizing Categorical Data**: Cat plots are great for showing the distribution of data points within different categories.
2. **Comparing Distributions**: By using the `hue`, `col`, and `row` parameters, you can compare the distributions of different subsets of the data within the same categories.
3. **Customizable Plots**: Cat plots offer a variety of plot types and customization options, making it easy to tailor visualizations to specific needs.

### Example Scenarios

1. **Comparing Tips Given on Different Days with Different Plot Types**:

   ```python

   ```

